```python
# 데이터 불러오기
import pandas as pd

bike_df = pd.read_csv('./datasets/bikerental.csv')
bike_df.head()
```

```python
# feature label 분리
features = bike_df.loc[:, ['season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed']]
label = bike_df["count"]
```
